````python
# Doubly Linked List

class DoublyLinkedList:
    class __Node:
        def __init__(self, data):
            self.data = data
            self.next = None
            self.prev = None  

    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0

    def append(self, value):
        self.size += 1
        new_node = self.__Node(value)

        if not self.head:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            new_node.prev = self.tail
            self.tail = new_node

    def insert(self, index, value):
        new_node = self.__Node(value)

        if index < 0 or index > self.size:
            raise IndexError("Index out of range")

        if index == 0:
            # Insert at the beginning
            new_node.next = self.head
            if self.head:
                self.head.prev = new_node
            self.head = new_node
            if not self.tail:
                self.tail = new_node
        else:
            current = self.head
            for _ in range(index - 1):
                current = current.next

            new_node.prev = current.prev
            new_node.next = current
            if current.prev:
                current.prev.next = new_node
            current.prev = new_node

            if not new_node.next:
                self.tail = new_node

        self.size += 1

    def remove(self, value):
        # Removes the first instance of "value" in the list
        current = self.head
        prev = None
        found = False

        while current and not found:
            if current.data == value:
                found = True
            else:
                prev = current
                current = current.next

        if found:
            if not prev:  # If the node to be removed is the head
                self.head = self.head.next
                if self.head:
                    self.head.prev = None
                else:
                    self.tail = None  # If the list becomes empty
            else:
                if current == self.tail:  # If the node to be removed is the tail
                    self.tail = prev
                    prev.next = None
                else:
                    prev.next = current.next
                    current.next.prev = prev

            self.size -= 1
        else:
            raise ValueError("Value not found")

    # def index(self, value):
    

    # def search(self, value):

    def __str__(self):
        out = "["
        current = self.head
        if current:
            out += f"{current.data}"
            current = current.next
        while current:
            out += f", {current.data}"
            current = current.next
        out += "]"
        return out

    @property
    def is_empty(self):
        return self.size == 0

    # def __getitem__(self, index):
    

    # def __setitem__(self, index, value):
    

    
    

        
````
